QUESTION 7


performing the t test to find the differnce in q values of two varients


null hypothesis: There is no  significant difference in the q values of two varients because considering that they are taken from same  population 

```
q1=var1['Q']
q2=var2['Q']
```

```
stats.ttest_ind(q1,q2)
```

since the p value is greater than 0.5 ,null hypothesis is rejected .So 
we can conclude that there is significant difference in the q values of two variants
